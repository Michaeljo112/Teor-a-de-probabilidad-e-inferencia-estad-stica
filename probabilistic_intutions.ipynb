```python
def sum_2_tice(trial):
  x = [i for i in range(1, trial+1)]
  y = []
  resultados = [1, 2, 3, 4, 5, 6]
  for i in range(1, trial+1):
      # Crear una suma aleatoria
      suma = random.choices(resultados, k=1)[0] + random.choices(resultados, k=1)[0]
      y.append(suma)
  return(x, y)
```

```python
x, y = sum_2_tice(10)
```

```python
# importing the required module
import matplotlib.pyplot as plt

# plotting the points
plt.plot(x, y)

# naming the x axis
plt.xlabel('Intento')
# naming the y axis
plt.ylabel('Suma de las caras de los dados')

# function to show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Obtener listas de lanzamientos
x1, y1 = sum_2_tice(10)
x2, y2 = sum_2_tice(30)
x3, y3 = sum_2_tice(100)
x4, y4 = sum_2_tice(300)

fig, axs = plt.subplots(1, 4, figsize=(20, 5), layout='constrained')

# Dibujamos una subfigura configurando los ejes
axs[0].plot(x1, y1)
axs[0].set_ylabel('Suma caras')
axs[0].set_yticks(range(1, 13, 1))
axs[0].set_xticks(range(1, max(x1)+1, 1))

axs[1].plot(x2, y2)
axs[1].set_yticks(range(1, 13, 1))

axs[2].plot(x3, y3)
axs[2].set_yticks(range(1, 13, 1))

axs[3].plot(x4, y4)
axs[3].set_yticks(range(1, 13, 1))

# Adjust vertical_spacing = 0.5 * axes_height
plt.subplots_adjust(hspace=0.5)

# Add text in figure coordinates
plt.figtext(0.5, -.1, 'Número de lanzamiento', ha='center', va='center')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Histograma de los lanzamientos

# Obtener listas de lanzamientos
x1, y1 = sum_2_tice(10)
x2, y2 = sum_2_tice(30)
x3, y3 = sum_2_tice(100)
x4, y4 = sum_2_tice(300)

# Datls del histograma
frecuencias1 = {key: y1.count(key) for key in range(2, 13, 1)}
frecuencias2 = {key: y2.count(key) for key in range(2, 13, 1)}
frecuencias3 = {key: y3.count(key) for key in range(2, 13, 1)}
frecuencias4 = {key: y4.count(key) for key in range(2, 13, 1)}

fig, axs = plt.subplots(4, 2, figsize=(20, 10), layout='constrained')

# Dibujamos una subfigura configurando los ejes
axs[0, 0].plot(x1, y1)
axs[0, 0].set_yticks(range(1, 13, 1))
axs[0, 0].set_xticks(range(1, max(x1)+1, 1))

axs[0, 1].bar(list(frecuencias1.keys()), list(frecuencias1.values()))
axs[0, 1].set_xticks(range(1, 13, 1))

# Gráfico de 30 lanzamientos
axs[1, 0].plot(x2, y2)
axs[1, 0].set_yticks(range(1, 13, 1))
axs[1, 0].set_xticks(range(1, max(x2)+1, 1))

axs[1, 1].bar(list(frecuencias2.keys()), list(frecuencias2.values()))
axs[1, 1].set_xticks(range(1, 13, 1))

# Gráfico de 100 lanzamientos
axs[2, 0].plot(x3, y3)
axs[2, 0].set_yticks(range(1, 13, 1))

axs[2, 1].bar(list(frecuencias3.keys()), list(frecuencias3.values()))
axs[2, 1].set_xticks(range(1, 13, 1))

# Gráfico de 300 lanzamientos
axs[3, 0].plot(x4, y4)
axs[3, 0].set_yticks(range(1, 13, 1))

axs[3, 1].bar(list(frecuencias4.keys()), list(frecuencias4.values()))
axs[3, 1].set_xticks(range(1, 13, 1))
```
